### Exercise 8: Transformer and Pipeline with scikitlearn for Imputation processing
In the previous exercise, we have decided to estimate the missing values using the mean estimator and fill missing
values with the same value. However the imputation is an estimation of the data with a defined strategy as
the median, the mean, variance, etc. Scikitlearn API provides an simple class to take care of missing values:


```python
# SimpleImputer. To use this class, we have to create a SimpleImputer instance by specifying the strategy
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(strategy="mean")
```

```python
# since the mean could be calculated only on numerical attributes, you need to create a copy of the data without the text attributes. You call this copy as a origin Data Num
# Then you have to fit the mean on your origin Data Num
Origin_Data_num= initialData.drop(columns= initialData.select_dtypes(include=["object"]).columns).copy()
my_imputer.fit(Origin_Data_num)
```

```python
#the fit() function stores the result into its statistic instance variable. At any moment you can reuse this variable to repair new missing values. You can display this variable: imputer.statistics
my_imputer.statistics_
```

```python
# finaly You use the imputer to transform the numerical data:
Transform_Data_num = my_imputer.transform(Origin_Data_num)
```

```python
df_Transform_Data_num=pd.DataFrame(Transform_Data_num, columns= Origin_Data_num.columns , index = Origin_Data_num.index )
```

### Exercise 9: to learn more
You can create a pipeline to automate the cleaning process using the pipe and imputer. Reshape your code to 1) create a pipeline to handle numerical data and an other pipeline to handle categorical data

```python
# SimpleImputer. To use this class, we have to create a SimpleImputer instance by specifying the strategy
from sklearn.impute import SimpleImputer
my_numerical_imputer = SimpleImputer(strategy="mean")
my_categorical_imputer = SimpleImputer(strategy="most_frequent")
Origin_Data_num= initialData.select_dtypes(include=["int64", "float64"]).copy()
Origin_Data_log= initialData.select_dtypes(include=["object"]).copy()
my_numerical_imputer.fit(Origin_Data_num)
my_categorical_imputer.fit(Origin_Data_log)
Transform_Data_num = my_numerical_imputer.transform(Origin_Data_num)
Transform_Data_log = my_categorical_imputer.transform(Origin_Data_log)
df_Transform_Data_num=pd.DataFrame(Transform_Data_num, columns= Origin_Data_num.columns , index = Origin_Data_num.index )
df_Transform_Data_log=pd.DataFrame(Transform_Data_log, columns= Origin_Data_log.columns , index = Origin_Data_log.index )
df_Transform_Data = pd.concat([df_Transform_Data_num, df_Transform_Data_log], axis=1)
df_Transform_Data
```
